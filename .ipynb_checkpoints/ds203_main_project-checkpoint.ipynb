```python
import numpy as np
import pandas as pd
import PIL
import matplotlib.pyplot as plt
```

```python
import nltk
nltk.download("stopwords")
nltk.download("punkt")
import string
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
from nltk import word_tokenize
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/sheelfshah/iitb_ds203_project/master/INvideos.csv")
df_img = pd.read_csv("https://raw.githubusercontent.com/sheelfshah/iitb_ds203_project/master/image_data/INpics/status.csv")
```

```python
df['img_status'] = df_img['0']
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
def text_preprocess(s):
  s=str(s)
  s = s.lower()
  s = "".join([char for char in s if char not in string.punctuation])
  s = word_tokenize(s)
  s = [word for word in s if word not in stop_words]
  s = [porter.stem(word) for word in s]
  return " ".join(s)
```

```python
df['description'] = df['description'].apply(text_preprocess)
```

```python
df['title'] = df['title'].apply(text_preprocess)
```

```python
df.drop(['tags'], axis=1, inplace=True)
```

```python
df["trend_date"]=df["trending_date"].apply(lambda s: s[-2:])
df["trend_month"]=df["trending_date"].apply(lambda s: s[-5:-3])
df["trend_year"]=df["trending_date"].apply(lambda s: s[:2])
df.drop("trending_date", axis=1, inplace=True)
```

```python
df.head()
```

```python
df["pub_date"]=df["publish_time"].apply(lambda s: s[8:10])
df["pub_month"]=df["publish_time"].apply(lambda s: s[5:7])
df["pub_year"]=df["publish_time"].apply(lambda s: s[2:4])
df["pub_hour"]=df["publish_time"].apply(lambda s: s[11:13])
df.drop("publish_time", axis=1, inplace=True)
```

```python
df.head()
```

```python
df['file_path'] = np.nan
df['file_path'] = df['file_path'].astype(str)

for i in range(df.shape[0]):
    if df.iloc[i]['img_status'] == True:
        df.at[i,'file_path'] = "https://raw.githubusercontent.com/sheelfshah/iitb_ds203_project/master/image_data/INpics/image-{}.jpg".format(i)

df.drop(df[df['img_status']==False].index,inplace=True)
df.drop(columns=['thumbnail_link','img_status'],inplace=True)
```

```python
print(df.shape)
display(df.head())
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 

title  = df['title']

comment_words = '' 
stopwords = set(STOPWORDS) 

for val in title: 
    
    val = str(val) 
    tokens = val.split() 

    comment_words += " ".join(tokens)+" "

wordcloud = WordCloud(width = 800, height = 800, 
                    background_color ='white', 
                    stopwords = stopwords, 
                    min_font_size = 10).generate(comment_words) 


plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show() 
```
