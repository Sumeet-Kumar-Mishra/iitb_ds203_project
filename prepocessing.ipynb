```python
import numpy as np
import pandas as pd
import PIL
import matplotlib.pyplot as plt
```

```python
import nltk
#nltk.download("stopwords")
#nltk.download("punkt")
import string
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
from nltk import word_tokenize
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
```

```python
df = pd.read_csv("csv_files/INvideos.csv")
df_img = pd.read_csv("https://raw.githubusercontent.com/sheelfshah/iitb_ds203_project/master/image_data/INpics/status.csv")

#change file path to the one in github
dfGB = pd.read_csv("csv_files/GBvideos.csv")
dfUS = pd.read_csv("csv_files/USvideos.csv")

dfs = [df,dfGB,dfUS]
```

```python
df['img_status'] = df_img['0']
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
def text_preprocess(s):
  s=str(s)
  s = s.lower()
  s = "".join([char for char in s if char not in string.punctuation])
  s = word_tokenize(s)
  s = [word for word in s if word not in stop_words]
  s = [porter.stem(word) for word in s]
  return " ".join(s)
```

```python
for data in dfs:
    data['description'] = data['description'].apply(text_preprocess)
```

```python
for data in dfs:
    data['title'] = data['title'].apply(text_preprocess)
```

```python
for data in dfs:
    data.drop(['tags'], axis=1, inplace=True)
```

```python
for data in dfs:
    data["trend_date"]=data["trending_date"].apply(lambda s: s[-2:])
    data["trend_month"]=data["trending_date"].apply(lambda s: s[-5:-3])
    data["trend_year"]=data["trending_date"].apply(lambda s: s[:2])
    data.drop("trending_date", axis=1, inplace=True)
```

```python
df.head()
```

```python
for data in dfs:
    data["pub_date"]=data["publish_time"].apply(lambda s: s[8:10])
    data["pub_month"]=data["publish_time"].apply(lambda s: s[5:7])
    data["pub_year"]=data["publish_time"].apply(lambda s: s[2:4])
    data["pub_hour"]=data["publish_time"].apply(lambda s: s[11:13])
    data.drop("publish_time", axis=1, inplace=True)
```

```python
#image path only added for df describing Indian data

df['file_path'] = np.nan
df['file_path'] = df['file_path'].astype(str)

for i in range(df.shape[0]):
    if df.iloc[i]['img_status'] == True:
        df.at[i,'file_path'] = "https://raw.githubusercontent.com/sheelfshah/iitb_ds203_project/master/image_data/INpics/image-{}.jpg".format(i)

df.drop(df[df['img_status']==False].index,inplace=True)
df.drop(columns=['thumbnail_link','img_status'],inplace=True)
```

```python
df.to_csv("csv_files/INpreprocessed.csv", index=False)
dfGB.to_csv("csv_files/GBpreprocessed.csv",index=False)
dfUS.to_csv("csv_files/USpreprocessed.csv",index=False)
print(df.shape)
display(df.head())
```
